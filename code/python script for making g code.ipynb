```python
def generate_gcode(grid_size, step_size, wait_time):
    gcode = [
        "G21 ; Set units to millimeters",
        "G90 ; Use absolute positioning",
        "G0 Z0 ; Move to the starting Z position",
        "M400 ; Wait for current moves to finish",
        f"G4 P{wait_time} ; Wait for {wait_time / 1000} seconds",
        "G0 X0 Y0 ; Move to the starting position",
    ]

    for y in range(0, grid_size + 1, step_size):
        if y % (2 * step_size) == 0:
            x_range = range(0, grid_size + 1, step_size)
        else:
            x_range = range(grid_size, -1, -1 * step_size)
        
        for x in x_range:
            gcode.append(f"G0 X{x} Y{y}")
            gcode.append("M400")
            gcode.append(f"G4 P{wait_time}")

    gcode.extend([
        "G0 X0 Y0 ; Return to the starting position",
        "M400 ; Wait for current moves to finish",
        f"G4 P{wait_time} ; Wait for {wait_time / 1000} seconds",
        "M30 ; End of program"
    ])

    return "\n".join(gcode)

# Example usage
grid_size = 40  # Size of the grid
step_size = 2   # Distance moved each time
wait_time = 1000  # Time between movements in milliseconds

print(generate_gcode(grid_size, step_size, wait_time))
```

```python
def generate_gcode(grid_size, move_up, wait_time, increment , x_cent ,y_cent, z_hight):
    grid_size = int(grid_size / 2)  # Grid size
    count=0

    gcode = []
    
    def add_wait():
        gcode.append(f"G4 P{wait_time}")
    
    def move_to(x, y, z):
        gcode.append(f"G0 X{x} Y{y} Z{z} F500")
        add_wait()

    
    for z in range(0, move_up+1, increment):
        if (z//increment) % 2 == 0:
            for y in range(-grid_size, grid_size+1, increment):
                if (y//increment) % 2 == 0:
                    for x in range(-grid_size, grid_size+1, increment):
                        move_to(x+x_cent, y+y_cent, z+z_hight)
                        
                else:
                    for x in range(grid_size, -grid_size-1, -increment):
                        move_to(x+x_cent, y+y_cent, z+z_hight)
                        print(x,y,z)
                        
        else:

            for y in range(grid_size, -grid_size-1, -increment):
                if (y//increment) % 2 == 0:    
                    for x in range(grid_size, -grid_size-1, -increment):
                        move_to(x+x_cent, y+y_cent, z+z_hight)
                        print(x,y,z)
                       
                else:
                    for x in range(-grid_size, grid_size+1, increment):
                        move_to(x+x_cent, y+y_cent, z+z_hight)
                        print(x,y,z)
                        

    
    print(count)
    
    
    return "\n".join(gcode)

# Parameters
grid_size = 40  # Grid size
move_up = 30  # Number of times to move up
wait_time = 5000  # Wait time in seconds
increment = 10  # Distance between each move
x_cent = 0
y_cent = 0
z_hight = 0
z_hight = z_hight +17.5

# Generate G-code
gcode = generate_gcode(grid_size, move_up, wait_time, increment, x_cent, y_cent, z_hight)

# Save to file
with open("generated_gcode.gcode", "w") as file:
    file.write(gcode)

print("G-code generated and saved to 'generated_gcode.gcode'")
```
